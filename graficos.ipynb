```python
import lammps_logfile
import matplotlib.pyplot as plt
```

```python
log = lammps_logfile.File("log.lammps")
x = log.get("Temp")
y3 = log.get("PotEng")

plt.plot(x, y3, label='PotEng')
plt.xlim(250,2500)
plt.xlabel('Temperatura (K)')
plt.ylabel('Energía (eV)')
plt.title('Curvas de Energía en función de la Temperatura')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
# Nombre del archivo de entrada y salida
input_file = "meltingV.dat"
temperature = []
V_Vo = []
# Abrir el archivo de entrada en modo lectura y el archivo de salida en modo escritura
# Temperature K Vol/Vo
with open(input_file, "r") as infile:
    for line in infile:
        if line.startswith('#'):
            continue
        data = line.strip().split()
        temperature.append(float(data[0]))
        V_Vo.append(float(data[1]))

       
# Crear la gráfica
#plt.figure(figsize=(10, 6))

plt.plot(temperature, V_Vo, marker = 'o', color='b')
plt.title('Vol/Vo vs. Temperature')
plt.xlabel('Temperature')
plt.ylabel('Vol/Vo')
plt.grid(True)
plt.show()
```
